```python
# import XGBoost
import xgboost as xgb


# define data_dmatrix
data_dmatrix = xgb.DMatrix(data=cust_x_xgb,label=cust_y_xgb) #converting entire x and y to dmatrix format
```

```python
# split cust_x_xgb and cust_y_xgb into training and testing sets

from sklearn.model_selection import train_test_split


x_train,x_test,y_train,y_test=train_test_split(cust_x_xgb,cust_y_xgb,test_size=0.20, random_state=42)
#Here we are using the split ratio of 80:20. The 20% testing data set is represented by the 0.2 at the end.

#this next step is optional to imporve performance
dtrain = xgb.DMatrix(x_train, label=y_train)
dtest = xgb.DMatrix(x_test, label=y_test)

#print(x_train.shape, x_test.shape)
#print(y_train.shape, y_test.shape)

#x_train
#x_test
#y_train
#y_test
```

```python
# import XGBClassifier
from xgboost import XGBClassifier

# declare parameters

params = {
            'objective':'binary:logistic', #both are predefined function # logistic regression function is being used
            'max_depth': 4,      #check by changing the value
            'learning_rate': 0.1,   #check by changing values between 0 to 1 # 1 means high variance model
            'n_estimators':100  # also check by changing number of estimators (10) # how many trees to be created
        }       

  
           
          
# instantiate the classifier 
xgb_clf = XGBClassifier(**params)





modelXG=xgb.train(params,dtrain)
preds = modelXG.predict(dtest)
```

```python
# make predictions on test data

#y_pred = xgb_clf.predict(X_test)
#y_pred
```

```python
# compute and print accuracy score

from sklearn.metrics import accuracy_score

#print(accuracy_score(y_test, preds ))


accuracy_score(y_test, preds.round(), normalize=True)
```
